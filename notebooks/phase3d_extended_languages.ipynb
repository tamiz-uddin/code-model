```python
# ============================================================
# Process Extended Languages Dataset
# ============================================================

import json
from pathlib import Path

def process_extended_dataset():
    """Process extended languages dataset."""
    print("Processing extended languages dataset...")
    
    # Load dataset
    input_file = Path("data/phase3d/data.jsonl")
    
    samples = []
    with open(input_file) as f:
        for line in f:
            samples.append(json.loads(line))
    
    print(f"✓ Loaded {len(samples)} samples")
    
    # Save processed
    output_dir = Path("data/phase3d/processed")
    output_dir.mkdir(parents=True, exist_ok=True)
    
    output_file = output_dir / "data.jsonl"
    with open(output_file, "w") as f:
        for sample in samples:
            f.write(json.dumps({"content": sample["content"]}) + "\n")
    
    print(f"✓ Saved to {output_file}")
    print(f"✓ File size: {output_file.stat().st_size / 1024:.1f} KB")

process_extended_dataset()
```

```python
# ============================================================
# PHASE 3D: Fine-tune on Extended Languages
# ============================================================

import sys
import torch
import yaml
import json
from pathlib import Path

sys.path.insert(0, str(Path.cwd()))

from models.architecture import CodeModel
from training.trainer import CodeModelTrainer

def train_phase3d():
    """Fine-tune Phase 3C model on extended languages."""
    print("=" * 60)
    print("PHASE 3D: EXTENDED LANGUAGES FINE-TUNING")
    print("=" * 60)
    
    # Check GPU
    print("\n1. Checking GPU...")
    print(f"   GPU Available: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"   GPU Name: {torch.cuda.get_device_name(0)}")
        print(f"   GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
    
    # Load config
    print("\n2. Loading config...")
    with open("config.yaml") as f:
        config = yaml.safe_load(f)
    
    # Optimize for Colab and fine-tuning
    config["training"]["batch_size"] = 16
    config["training"]["num_epochs"] = 1
    config["training"]["learning_rate"] = 1e-6  # Very low LR for fine-tuning
    config["training"]["warmup_steps"] = 5
    
    print(f"   Batch size: {config['training']['batch_size']}")
    print(f"   Learning rate: {config['training']['learning_rate']} (very low for fine-tuning)")
    print(f"   Epochs: {config['training']['num_epochs']}")
    
    # Create model
    print("\n3. Creating model...")
    model = CodeModel(config["model"]).get_model()
    total_params = sum(p.numel() for p in model.parameters())
    print(f"   ✓ Model created")
    print(f"   Parameters: {total_params:,}")
    
    # Load Phase 3C checkpoint
    print("\n4. Loading Phase 3C checkpoint...")
    checkpoint_path = Path("outputs/models/fullstack-model-v1.pt")
    
    if checkpoint_path.exists():
        model.load_state_dict(torch.load(checkpoint_path))
        print(f"   ✓ Loaded checkpoint: {checkpoint_path}")
        print(f"   Knowledge: JS + Python + HTML + CSS + Frameworks + Databases + APIs")
    else:
        print(f"   ⚠ Checkpoint not found: {checkpoint_path}")
        print("   Training from scratch (not recommended)")
    
    # Load data
    print("\n5. Loading dataset...")
    data_path = Path("data/phase3d/processed/data.jsonl")
    
    samples = []
    with open(data_path) as f:
        for line in f:
            samples.append(json.loads(line))
    
    print(f"   ✓ Loaded {len(samples)} samples")
    print(f"   Languages: TypeScript, Java, Go, Rust, C#, SQL")
    
    # Fine-tune
    print("\n6. Starting fine-tuning...")
    print("   (Very low learning rate preserves all prior knowledge)")
    
    trainer = CodeModelTrainer(model, config, samples)
    trainer.train()
    
    # Save model
    print("\n7. Saving fine-tuned model...")
    output_dir = Path("outputs/models")
    output_dir.mkdir(parents=True, exist_ok=True)
    
    model_path = output_dir / "universal-model-v1.pt"
    torch.save(model.state_dict(), model_path)
    print(f"   ✓ Saved to {model_path}")
    print(f"   Size: {model_path.stat().st_size / 1e6:.1f} MB")
    
    print("\n" + "=" * 60)
    print("✅ PHASE 3D FINE-TUNING COMPLETE")
    print("=" * 60)
    print(f"\nUniversal Model ready: {model_path}")
    print("Knowledge:")
    print("  - Frontend: JavaScript, TypeScript, HTML, CSS, React, Vue, Angular")
    print("  - Backend: Python, Express, Django, Flask")
    print("  - Systems: Java, Go, Rust, C#")
    print("  - Data: SQL, MongoDB")
    print("  - APIs: REST, GraphQL, JWT")
    print("\n🎉 UNIVERSAL CODING MODEL COMPLETE!")
    print("Next: Evaluation and Deployment")
    
    return True

train_phase3d()
```
